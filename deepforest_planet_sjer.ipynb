```python
from deepforest import deepforest
from matplotlib import pyplot as plt
import rasterio as rio
from deepforest import get_data
from rasterio.windows import Window
import numpy as np
from earthpy.spatial import bytescale
```

```python
test_model = deepforest.deepforest()
test_model.use_release()
```

```python
test_wv2 = r"D:\projects\RD\planet\SPIE_research\september_2017_planet_scenes\20170902_180146_1029\20170902_180146_1029_3B_AnalyticMS_SR.tif"
with rio.open(test_wv2) as src:
    print(src.profile)
```

```python
# get some arbitrary window
min_col = 5500
min_row = 2000
num = 512
with rio.open(test_wv2) as src:
    w = src.read(window=Window(min_col, min_row, num, num))
    w = w/10000 #using surface reflectance so divide by 10000
    
print(w.min(), w.max())
plt.imshow(w[3])
```

```python
raw_arr = np.rollaxis(w[[0,1,2], :, :], 0, 3) #get images into b-g-r format. Planet has b-g-r-nir channel spec
raw_arr = bytescale(raw_arr) # make to range 0-255
pred = test_model.predict_image(raw_image=raw_arr, color=[0,0,255])
```

```python
fig, axes = plt.subplots(1,3,figsize=(15,10))
for i in range(3):
    axes[i].imshow(pred[:,:,i])
```

```python
plt.figure(figsize=(15,15))
plt.imshow(pred[...,::-1])
```

```python
from deepforest import preprocess
```

```python
windows = preprocess.compute_windows(raw_arr, patch_size=20,patch_overlap=0.1)
```

```python
index=0
crop = raw_arr[windows[index].indices()]
```

```python
prediction = test_model.predict_image(raw_image = crop,return_plot=True, color=[0,0,255])
```

```python
plt.figure(figsize=(15,15))
plt.imshow(prediction[...,::-1])
```
